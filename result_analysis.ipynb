```python
def plot_prediction_truth(file_add, title, num):
    test_df = pd.read_csv(file_add, sep=',', names = ['vol','ob'])
    
    fig, ax = plt.subplots()
    fig.set_size_inches( 18,7 )
    
    ax.plot(test_df.iloc[:num], lw=2)
    ax.legend( ['Prediction','True values'], loc=0)
    ax.set_ylim(-.1, 1.5)
    ax.legend()
    
    plt.title(title)
```

```python
num_toshow = 500
plot_prediction_truth('res/pytest_bayes.txt', 'Bayes regression', num_toshow)
plot_prediction_truth('res/pytest_enet.txt', 'Elastic net', num_toshow)
plot_prediction_truth('res/pytest_rf.txt', 'Random forest', num_toshow)
plot_prediction_truth('res/pytest_gbt.txt', 'Gradient boostet tree', num_toshow)

plot_prediction_truth('res/pytest.txt', 'Mixture linear', num_toshow)
```
